```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('cost_revenue_dirty.csv')
```

```python
# description of the data set.
df.describe()
```

```python
#seperate the data for prediction
# Using linear Regression
# X = production budget
# y = world wide gross revenue

X = DataFrame(df, columns=['production_budget_usd'])
y = DataFrame(df, columns=['worldwide_gross_usd'])
```

```python

df.describe()
```

```python
plt.figure(figsize=(15,9))
plt.scatter(X, y, alpha = 0.2)
plt.title( 'Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

# Run and Evaluation of Linear Reggression

```python
regression = LinearRegression()
regression.fit(X, y)
```

# Slope Coefficient

```python
regression.coef_ # theta_1
```

# Intercept

```python
regression.intercept_ #
```

# Plotting coeff. and intercept with data

```python
plt.figure(figsize=(15,9))

plt.scatter(X, y, alpha = 0.2)
plt.plot(X, regression.predict(X), color = 'red', linewidth = 3)

plt.title( 'Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')

plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

# We have slope coefficient 3.11150918. Which is a positive relation with budget and revenue. More precisely it can be interpreted for each dollar spent for production we get revenue of 3.11150918 dollars

# Also, we have intercept of -7236192.72913851. Which indicates the movies of budget of 0 dollar loose approximate -7236192 dollars. Which seems unrealistic.

# Goodness of fit (R^2) can be used to predict more accurately.

# Finding R^2
regression.score(X, y)
